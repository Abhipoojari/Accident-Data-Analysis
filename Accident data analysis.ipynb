```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("NCRB_ADSI-2020_Table_1A.9.csv")

# Display basic info and first few rows
display(df.info())
display(df.head())

# Data Cleaning & Preprocessing
# Handling missing values
df.fillna(0, inplace=True)  # Replace NaN values with 0 for simplicity

# Standardizing column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Identifying areas with the highest and lowest number of accidents
accident_summary = df.groupby('area_name')['total_accidents'].sum().reset_index()

# Defining safe and unsafe areas (Assuming lower accidents = safer)
thresh = accident_summary['total_accidents'].median()
safe_areas = accident_summary[accident_summary['total_accidents'] <= thresh]
unsafe_areas = accident_summary[accident_summary['total_accidents'] > thresh]

# Exploratory Data Analysis (EDA)
plt.figure(figsize=(12,6))
sns.histplot(df['total_accidents'], bins=30, kde=True, color='blue')
plt.xlabel('Total Accidents')
plt.ylabel('Frequency')
plt.title('Distribution of Total Accidents')
plt.show()

# Visualization - Top 10 Unsafe Areas
top_unsafe = unsafe_areas.nlargest(10, 'total_accidents')
plt.figure(figsize=(12,6))
sns.barplot(x='total_accidents', y='area_name', data=top_unsafe, palette='Reds_r')
plt.xlabel('Total Accidents')
plt.ylabel('Area Name')
plt.title('Top 10 Unsafe Areas for Driving')
plt.show()

# Visualization - Top 10 Safe Areas
top_safe = safe_areas.nsmallest(10, 'total_accidents')
plt.figure(figsize=(12,6))
sns.barplot(x='total_accidents', y='area_name', data=top_safe, palette='Greens_r')
plt.xlabel('Total Accidents')
plt.ylabel('Area Name')
plt.title('Top 10 Safe Areas for Driving')
plt.show()

# Additional Analysis
# Accident Causes Breakdown
cause_columns = [
    'dangerous_or_careless_driving/_over-taking/etc._-_cases',
    'over_speeding_-_cases',
    'driving_under_influence_of_drug/alcohol_-_cases',
    'physical_fatigue_of_drivers_-_cases',
    'defect_in_mechanical_condition_of_motor_vehicle_-_cases',
    'animal_crossing_-_cases',
    'weather_condition_(total)_-_cases',
    'lack_of_road_infrastructure_-_cases',
    'vehicles_parking_at_road_shoulders_-_cases',
    'causes_not_known_-_cases',
    'other_causes_-_cases'
]

cause_summary = df[cause_columns].sum().sort_values(ascending=False)

# Accident Severity by Cause
severity_columns = {
    'Dangerous Driving': ['dangerous_or_careless_driving/_over-taking/etc._-_injured',
                          'dangerous_or_careless_driving/_over-taking/etc._-_died'],
    'Over Speeding': ['over_speeding_-_injured', 'over_speeding_-_died'],
    'Drunk Driving': ['driving_under_influence_of_drug/alcohol_-_injured',
                      'driving_under_influence_of_drug/alcohol_-_died'],
    'Fatigue': ['physical_fatigue_of_drivers_-_injured', 'physical_fatigue_of_drivers_-_died'],
    'Mechanical Defects': ['defect_in_mechanical_condition_of_motor_vehicle_-_injured',
                           'defect_in_mechanical_condition_of_motor_vehicle_-_died'],
    'Weather Conditions': ['weather_condition_(total)_-_injured', 'weather_condition_(total)_-_died'],
}

severity_data = {cause: df[columns].sum().values for cause, columns in severity_columns.items()}
severity_df = pd.DataFrame(severity_data, index=['Injuries', 'Fatalities']).T

# Road Infrastructure Related Accidents
infra_accidents = df[['lack_of_road_infrastructure_-_cases']].sum()

# Drunk Driving Analysis
drunk_driving_cases = df[['driving_under_influence_of_drug/alcohol_-_cases']].sum()

# Unmanned Railway Crossing Accidents
railway_crossing_accidents = df[['unmanned_railway_crossing_accidents_-_cases']].sum()

# Visualization
plt.figure(figsize=(12,6))
sns.barplot(x=cause_summary.values, y=cause_summary.index, palette="Blues_r")
plt.xlabel("Number of Cases")
plt.ylabel("Accident Cause")
plt.title("Top Causes of Road Accidents")
plt.show()

severity_df.plot(kind='bar', stacked=True, figsize=(12,6), colormap="coolwarm")
plt.xlabel("Accident Cause")
plt.ylabel("Number of Cases")
plt.title("Accident Severity by Cause (Injuries vs. Fatalities)")
plt.xticks(rotation=45)
plt.legend(title="Severity Type")
plt.show()

fatality_data = severity_df["Fatalities"].sort_values(ascending=False)
plt.figure(figsize=(8,8))
plt.pie(fatality_data, labels=fatality_data.index, autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Fatality Distribution by Accident Cause")
plt.show()

# Summary Output
display("Safe Areas:", safe_areas.head(10))
display("Unsafe Areas:", unsafe_areas.head(10))
display("Top Causes of Accidents:", cause_summary)
display("Accident Severity Data:", severity_df)
display("Road Infrastructure Accidents:", infra_accidents)
display("Drunk Driving Cases:", drunk_driving_cases)
display("Unmanned Railway Crossing Accidents:", railway_crossing_accidents)
```
